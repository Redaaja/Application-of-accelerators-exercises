```python
class Nuclide:
    def __init__(self, Z, A):
        self.Z = Z  # Atomic_number
        self.A = A  # number_of_nucleons

    def get_neutrons(self):
        return self.A - self.Z

example_nuclide = Nuclide(Z=52, A=131)

neutrons_number = example_nuclide.get_neutrons()

print(f"The number of neutrons for the example nuclide is: {neutrons_number}")
```
