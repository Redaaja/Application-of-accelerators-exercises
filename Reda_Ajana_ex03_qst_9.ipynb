```python
import numpy as np
import matplotlib.pyplot as plt
#importing the data
file = np.genfromtxt('2016-07-11_ipm_data.txt')
#plot the data
plt.plot(file)
plt.grid(color='black')
plt.xlabel('position')
plt.ylabel('signal')
plt.title('ionisation beam')
y = max(file)
x = np.where(file == y)
plt.plot(x,y, 'rx')
plt.savefig('Reda_Ajana_plot')
```
